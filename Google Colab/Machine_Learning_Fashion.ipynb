```python
import matplotlib.pyplot as pyplot
import pandas as pd
import numpy as np
import tensorflow as tf  # Import TensorFlow library
from tensorflow import keras  # Import Keras module from TensorFlow
import matplotlib.pyplot as plt

fashion_mnist = keras.datasets.fashion_mnist  # Load Fashion MNIST dataset from Keras

# Load the Fashion MNIST dataset into training and testing sets
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist.load_data()

# Print the shape and data type of the training set
print(X_train_full.shape, X_train_full.dtype)

# Print the shape and data type of the labels in the training set
print(y_train_full.shape, y_train_full.dtype)

# Print the shape and data type of the testing set
print(X_test.shape, X_test.dtype)

# Print the shape and data type of the labels in the testing set
print(y_test.shape, y_test.dtype)

# Define class names corresponding to the integer labels
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
               "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
print(" ")

# Print the mapping of class labels to class names
for label, class_name in enumerate(class_names):
    print(f"Label {label}: {class_name}")

print(" ")

# Display the first 9 images in the training set
for i in range(9):
    # Create subplots with 3 rows and 3 columns
    pyplot.subplot(330 + 1 + i)

    # Show the current image in grayscale
    pyplot.imshow(X_train_full[i], cmap=pyplot.get_cmap('gray'))

# Show the plot with all 9 images
pyplot.show()

# Split the dataset into validation and training sets
# Reserve 5000 samples for validation, rest for training
X_valid, X_train = X_train_full[:5000] / 255.0, X_train_full[5000:] / 255.0

# Split labels accordingly
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]

# Create a Sequential model
model = keras.models.Sequential()

# Flatten layer to convert each 28x28 image into a 1D array
model.add(keras.layers.Flatten(input_shape=[28, 28]))

# Dense layer with 300 neurons and ReLU activation
model.add(keras.layers.Dense(300, activation="relu"))

# Dense layer with 100 neurons and ReLU activation
model.add(keras.layers.Dense(100, activation="relu"))

# Output Dense layer with 10 neurons (one per class) and softmax activation
model.add(keras.layers.Dense(10, activation="softmax"))

# Display model architecture and parameters
model.summary()

# Get the layers of the model
my_layers = model.layers

# Select the first hidden layer
first_hidden_layer = my_layers[1]

# Extract the weights and biases from the first hidden layer
weights, biases = first_hidden_layer.get_weights()

# Compile the model
model.compile(
    # Loss function for integer labels
    loss="sparse_categorical_crossentropy",
    # Stochastic Gradient Descent optimizer
    optimizer="sgd",
    # Metrics for model evaluation
    metrics=["accuracy"]
)

# Train the model
history = model.fit(
    X_train,            # Training features
    y_train,            # Training labels
    epochs=30,          # Number of epochs
    validation_data=(X_valid, y_valid)  # Validation set
)

# Training history
history_dict = history.history

# -------------------------------
# Loss plot
# -------------------------------
plt.plot(history_dict['loss'], label='Training Loss')
plt.plot(history_dict['val_loss'], label='Validation Loss')
plt.title('Loss per Epoch')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

# -------------------------------
# Accuracy plot
# -------------------------------
plt.plot(history_dict['accuracy'], label='Training Accuracy')
plt.plot(history_dict['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy per Epoch')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# Evaluate model on test set
model.evaluate(X_test, y_test)

# Select new instances for prediction
X_new = X_test[:3]

# Predict probabilities using the trained model
y_proba = model.predict(X_new)

# Print predicted probabilities rounded to two decimals
print(y_proba.round(2))

# Get predicted class indices
class_names_idx = np.argmax(y_proba, axis=1)

# Map predicted indices to class names
result = [class_names[i] for i in class_names_idx]

# Print predicted class names
print(result)
```
